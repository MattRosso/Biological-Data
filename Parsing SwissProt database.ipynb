```python
import gzip
from Bio import SeqIO
```

```python
# Get GO and taxonomy IDs
path = "insert_path_to_database_here.gz" 

# Entire SwissProt could be accessed at: ftp://ftp.uniprot.org/pub/databases/uniprot/current release/knowledgebase/complete/uniprot sprot.xml.gz
# Circa 700 Mb, circa 15 minutes of parsing

seq_records = {}

i = 0
handle = gzip.open(path)
for record in SeqIO.parse(handle, "uniprot-xml"):
    name = record.name
    description = record.description
    GO_list = [x[3:] for x in record.dbxrefs if x[0:3] == 'GO:']
    Taxonomy_id = [x[14:] for x in record.dbxrefs if x[0:14] == "NCBI Taxonomy:"]
    if not Taxonomy_id:
      print(record.name, "no taxonomy id found")
    seq_records[record.id] = (name, description, GO_list, Taxonomy_id)
    i = i+1
    if i%5000 == 0:
      print(i)
```

```python
import pickle

path = "data/go_swissprot_taxonomy.pickle"

with open(path, 'wb') as handle:
    pickle.dump(seq_records, handle)
```
